```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Charger le modèle et le tokenizer pour T5
tokenizer = AutoTokenizer.from_pretrained("google/t5-v1_1-base")  # Modèle T5
model = AutoModelForSeq2SeqLM.from_pretrained("google/t5-v1_1-base")

# Fonction de traduction avec préfixe T5
def translate_text_t5(text, source_lang="en", target_lang="ar"):
    if pd.isna(text) or not isinstance(text, str):  # Gérer les valeurs nulles
        return ""
    # Préfixe de la tâche pour T5
    task_prefix = f"translate {source_lang} to {target_lang}: "
    input_text = task_prefix + text
    # Encoder le texte
    inputs = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)
    # Générer la traduction
    outputs = model.generate(inputs, max_length=512, num_beams=4, early_stopping=True)
    # Décoder le texte traduit
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Charger le fichier CSV
input_file = "cleaned_ml_questions.csv"  # Remplacez par le chemin de votre fichier d'entrée
data = pd.read_csv(input_file)

# Vérifier les colonnes disponibles
if "question" not in data.columns or "answer" not in data.columns:
    raise ValueError("Les colonnes 'question' et 'answer' sont introuvables dans le fichier CSV.")

# Appliquer la traduction sur les colonnes ciblées
print("Traduction en cours...")
data["questions_ar"] = data["question"].apply(lambda x: translate_text_t5(x, source_lang="en", target_lang="ar"))
data["answers_ar"] = data["answer"].apply(lambda x: translate_text_t5(x, source_lang="en", target_lang="ar"))

# Sauvegarder le fichier traduit
output_file = "base_traduite_t5.csv"
data.to_csv(output_file, index=False, encoding="utf-8-sig")
print(f"Traduction terminée ! Le fichier traduit est enregistré sous '{output_file}'.")
```
